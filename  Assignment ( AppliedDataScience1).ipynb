```python
def generate_suicide_timeline(df):
    """
    Creates a line chart showing global suicide totals over time.
    The comments explain what's happening in each step.
    """
    
    # First group the data by year and sum the suicides
    print("Grouping the data by year...")
    yearly_data = df.groupby('year')['suicides/100k pop'].sum().reset_index()
    
    # Now let's visualize this data in a line chart 
    print("Creating the line chart...")   
    plt.figure(figsize=(10, 6))
    plt.plot(yearly_data['year'],yearly_data['suicides/100k pop'], marker='o', color='red')

    # Add some details to the chart to make it clearer
    print("Adding labels and styles...")
    plt.title("Worldwide Suicide Totals Per  100k pop ", fontsize=20)  
    plt.xlabel('Year', fontsize=16)
    plt.ylabel('Total Suicides 100k pop', fontsize=16)
    plt.grid(True)
    plt.xticks(rotation=30)
    plt.tight_layout()
    
    # Finally, display the chart
    print("Showing the chart...")
    plt.show()
    
    print("All done!")

# Example usage
generate_suicide_timeline(df)
```

```python
# Group df by age group and sum the suicides for all countries for the year 1987
global_1987_suicides_by_age_all_countries = df[df['year'] == 1987].groupby('age')['suicides_no'].sum().reset_index()

# Sort the age groups in a meaningful order
global_1987_suicides_by_age_all_countries['age'] = pd.Categorical(global_1987_suicides_by_age_all_countries['age'], ordered=True)
global_1987_suicides_by_age_all_countries = global_1987_suicides_by_age_all_countries.sort_values('age')

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(global_1987_suicides_by_age_all_countries['age'], global_1987_suicides_by_age_all_countries['suicides_no'], color='blue')
plt.title('Total Global Suicides by Age Group, 1987', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Suicides', fontsize=12)
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()  # Adjusts the plot to ensure everything fits without overlapping
plt.show()
```

```python
# Group df by sex and sum the suicides for 1987 for all countries
global_1987_suicides_by_sex_all_countries = df[df['year'] == 1987].groupby('sex')['suicides_no'].sum().reset_index()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(global_1987_suicides_by_sex_all_countries['suicides_no'], labels=global_1987_suicides_by_sex_all_countries['sex'], 
        colors=['cyan', 'magenta'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Global Suicides by Gender, 1987', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
